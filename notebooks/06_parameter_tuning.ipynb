```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from src.dbscan_from_scratch import DBSCAN
from src.parameter_tuning import ParameterSelector
from src.visualization import DBSCANVisualizer
from src.data_loader import DatasetGenerator
```

## 1. K-Distance Graph Method

The k-distance graph is the primary method recommended in the paper for selecting epsilon.

```python
# Generate sample data
generator = DatasetGenerator(random_state=42)
X = generator.generate_basic_shapes('moons', n_samples=300, noise=0.05)

# Create parameter selector
selector = ParameterSelector()

# Plot k-distance graph
viz = DBSCANVisualizer()
viz.plot_k_distance_graph(X, k=5, show_elbow=True)
plt.show()
```

## 2. Parameter Sensitivity Analysis

Explore how different parameter values affect clustering results.

```python
# Test different epsilon values
eps_values = [0.1, 0.2, 0.3, 0.4]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for idx, eps in enumerate(eps_values):
    dbscan = DBSCAN(eps=eps, min_pts=5)
    labels = dbscan.fit_predict(X)
    
    ax = axes[idx]
    viz.plot_clusters(X, labels, ax=ax, title=f'eps={eps}')

plt.tight_layout()
plt.show()
```

## 3. Automated Parameter Suggestion

Use the ParameterSelector to get automated suggestions.

```python
# Get suggested parameters
suggested_params = selector.suggest_parameters(X)
print(f"Suggested eps: {suggested_params['eps']:.3f}")
print(f"Suggested min_pts: {suggested_params['min_pts']}")

# Apply suggested parameters
dbscan = DBSCAN(eps=suggested_params['eps'], min_pts=suggested_params['min_pts'])
labels = dbscan.fit_predict(X)

viz.plot_clusters(X, labels, title='Clustering with Suggested Parameters')
plt.show()
```

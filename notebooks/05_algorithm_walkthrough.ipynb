## 2. Algorithm Overview [Paper §4.2, p. 228]

The DBSCAN algorithm works by:
1. Visiting each point sequentially
2. Finding its ε-neighborhood
3. If |neighbors| >= MinPts: create/expand cluster
4. Otherwise: mark as noise

### Key Operations
- **regionQuery(P, ε)**: Find N_ε(P) = {Q ∈ D | dist(P, Q) ≤ ε}
- **expandCluster**: Add all density-reachable points to cluster

## 3. Step-by-Step Execution

Let's create a simple dataset and trace the algorithm:

```python
# Create sample dataset
X_sample = np.array([
    [0, 0], [0.3, 0], [0.6, 0], [0.9, 0], [1.2, 0],
    [3, 0],  # isolated point
    [0, 1], [0.3, 1], [0.6, 1]
])

eps = 0.4
min_pts = 3

# Visualize
plt.figure(figsize=(12, 6))
plt.scatter(X_sample[:, 0], X_sample[:, 1], c='gray', s=200)
for i in range(len(X_sample)):
    plt.text(X_sample[i, 0], X_sample[i, 1] + 0.1, str(i), ha='center', fontsize=12)
plt.title(f'Sample Dataset (ε={eps}, MinPts={min_pts})')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.show()

# Run DBSCAN
dbscan = DBSCAN(eps=eps, min_pts=min_pts)
labels = dbscan.fit_predict(X_sample)
core_indices = dbscan.get_core_points()

# Display results
print('\nClustering Results:')
for i in range(len(X_sample)):
    point_type = dbscan.get_point_type(i).name
    label_str = 'NOISE' if labels[i] == -1 else f'Cluster {labels[i]}'
    print(f'Point {i}: {label_str:12s} ({point_type})')

viz.plot_point_types(X_sample, labels, core_indices, eps)
plt.show()
```

## 5. Summary

### Key Takeaways

1. **Algorithm Structure** [Paper §4.2, p. 228]:
   - Sequential point processing
   - Core point detection via regionQuery
   - Cluster expansion via density-reachability
   - Single-pass, deterministic execution

2. **Core Operations**:
   - **regionQuery(P, ε)**: Find N_ε(P) in O(n) time
   - **expandCluster**: Grow cluster from core point
   - **Point classification**: Core, border, or noise

3. **Complexity** [Paper §6]:
   - Time: O(n²) without spatial indexing
   - Time: O(n log n) with R-tree/KD-tree
   - Space: O(n) for labels and queues
